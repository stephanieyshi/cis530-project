```
from gensim.models.fasttext import FastText

# load model
model = FastText.load_fasttext_format('model_breitbart.bin')
```

```
# read in words
import numpy as np
import json

def get_words(filename):
    with open(filename, 'r') as f:
        lines = f.read().strip().split('\n')

    words = set()

    for line in lines:
        words.add(line)

    return list(words)

def read_json(filename):
    with open(filename, 'r') as f:
        return json.load(f)
```

```
i = 0
with open('breitbart_embedding_dict.txt', 'w+') as f:
    for word in model.wv.vocab:
        vec = model.wv[word]
        f.write(word)
        f.write(' ')
        for c in vec:
            f.write(str(c))
            f.write(' ')
        f.write('\n')
        i += 1
        if (i % 1000 == 0):
            print(i)
```
